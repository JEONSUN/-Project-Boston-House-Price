# Linear Regression Model

MEDV와의 상관계수가 -0.5 이하이거나 0.5 이상인 변수들(**LSTAT,PTRATIO,RM**)로 MEDV와의 선형 회귀를 구해보고자 한다.

```python
# x 와 y 분리하기

prices = data['MEDV']
features = data.drop('MEDV', axis = 1)
```

```python
features.head()
```

```python
prices.head()
```

```python
mean_price = np.mean(prices)
mean_price
```

```python
min_price = np.min(prices)
min_price
```

```python
max_price = np.max(prices)
max_price
```

```python
X = pd.DataFrame(np.c_[data['LSTAT'], data['RM'],data['PTRATIO']], columns = ['LSTAT','RM','PTRATIO'])
Y = data['MEDV']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 선형회귀모델을 만들어 lin_model에 저장
lin_model = LinearRegression()

# 모델을 학습하도록 함.
# x는 2차원 y는 기존의 값을 입력함.
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for training set
from sklearn.metrics import r2_score
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

**RMSE** : 추정 값 또는 모델이 예측한 값과 실제 환경에서 관찰되는 값의 차이를 다룰 때 흔히 사용하는 측도

**R2** : 결정계수는 추정한 선형 모형이 주어진 자료에 적합한 정도를 재는 척도 

```python
# rm과 medv의 회귀계수

lin_model.fit(X , y = data["MEDV"])

prediction = lin_model.predict(X)
print('a value = ', lin_model.intercept_)

print('b balue = ', lin_model.coef_)
```

```python
rm_X = pd.DataFrame(data['RM'], columns = ['RM'])
rm_Y = data['MEDV']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 선형회귀모델을 만들어 lin_model에 저장
lin_model = LinearRegression()

# 모델을 학습하도록 함.
# x는 2차원 y는 기존의 값을 입력함.
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for training set
from sklearn.metrics import r2_score
y_train_predict = lin_model.predict(X_train)
rm_rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
rm_r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rm_rmse))
print('R2 score is {}'.format(rm_r2))
print("\n")

# model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rm_rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
rm_rm_r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rm_rmse))
print('R2 score is {}'.format(rm_r2))
```
